```python
import numpy as np
import pandas as pd
```

```python
movies=pd.read_csv('tmdb_5000_movies.csv')
movies.info()
```

```python
credits=pd.read_csv('tmdb_5000_credits.csv')
credits.info()
```

```python
#merge the 2 datasets using uniqu emovie id and the movie title
```

```python
movies=movies.merge(credits,on='title')

```

```python
movies.shape
```

```python
credits.shape
```

```python
movies.info()
```

```python
movies=movies[['id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.isnull().sum()
movies.dropna()
movies.isnull().sum()
```

```python
movies.iloc[0].genres
import ast
```

```python
def convert(obj):
  L=[]
  for i in ast.literal_eval(obj):
    L.append(i['name'])
  return L
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies.head()
movies.info()
```

```python
movies['cast'][0] #name is the key we have to look for now and top 3 actors we looked
```

```python
def convert3(obj):
  L=[]
  counter=0
  for i in ast.literal_eval(obj):
    if counter!=3:
      L.append(i['name'])
      counter+=1
    else:
      break
  return L
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
movies['keywords']=movies['keywords'].apply(convert)
movies.head()
#we need to find dirrector in crew
```

```python
def find_Director(obj):
  L=[]
  for i in ast.literal_eval(obj):
    if i['job']=='Director':
      L.append(i['name'])
      break
  return L
```

```python
movies['crew']=movies['crew'].apply(find_Director)
```

```python
movies.head()
```

```python
movies['overview'][0]
movies['overview']=movies['overview'].fillna('').apply(lambda x:x.split())
movies.head()
```

```python
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
movies.head()
```

```python
movies['tags']=movies['genres']+movies['keywords']+movies['cast']+movies['crew']+movies['overview']
movies.head()
```

```python
movies=movies[['id','title','tags']]
movies.head()
```

```python
#now the tags are list but we need them in string
movies['tags']=movies['tags'].apply(lambda x:" ".join(x))
```

```python
movies['tags']=movies['tags'].apply(lambda x:x.lower())
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
vectors=cv.fit_transform(movies['tags']).toarray()
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
  y=[]
  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)
```

```python
movies['tags']=movies['tags'].apply(stem)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vectors)
cosine_similarity(vectors).shape
```

```python
list(enumerate(similarity))
```

```python
def recommend(user_provided_title):
   movie_index=movies[movies['title']==user_provided_title].index[0]
   distances=similarity[movie_index]
   movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
   for i in movies_list:
     print(movies.iloc[i[0]].title)
recommend('Batman Begins')
```
